데이터 완결성(completeness): 결측값이 없는 것 (ex)NaN -> Not a Number)  
데이터 유일성(uniqueness): 동일 데이터가 불필요하게 중복되면 안 됨  
데이터 통일성(conformity): 데이터가 동일한 형식으로 저장되어야 함  
데이터 정확성(accuracy): 데이터가 정확해야 함 (이상점(outlier) - > (다른 값들과 너무 동떨어져 있는 데이터)이 없도록)

```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# data cleaning: completeness
df = pd.read_csv('../data/week4/attendance.csv', index_col = 0)
df
```

```python
df.dropna(axis = 'columns') # axis로 index 기준으로 없앨 지, column 기준으로 없앨 지 결정
```

```python
df.mean()
```

```python
df.fillna(df.median()) #결측값을 중앙값으로 대체
```

```python
# data cleaning: accuracy
## +-1.5*IQR을 넘어가면 이상점으로 판단, 무조건 제거하지는 않음
df = pd.read_csv('../data/week4/beer.csv', index_col = 0)
df.head()
```

```python
df.plot(kind = 'box', y = 'abv')
```

```python
df['abv'].describe() # 25, 75% 확인
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

```python
iqr_condition = (df['abv'] < q1 - 1.5*iqr) | (df['abv'] > q3 + 1.5*iqr)
```

```python
df[iqr_condition]
```

```python
df.loc[2250, 'abv'] = 0.055 # 2250줄 데이터 수정

iqr_condition = (df['abv'] < q1 - 1.5*iqr) | (df['abv'] > q3 + 1.5*iqr) # 2250은 수정됐으므로 재선언

df.drop(df[iqr_condition].index, axis = 'index', inplace = True)
df
```

```python
# 수정 확인
iqr_condition = (df['abv'] < q1 - 1.5*iqr) | (df['abv'] > q3 + 1.5*iqr)
print(df[iqr_condition])
print(df.plot(kind = 'box', y = 'abv'))
```

```python
# 관계적 이상점 (Relational Outlier) -> 변수의 관계를 고려했을 때 이상한 데이터

df = pd.read_csv('../data/week4/exam_outlier.csv')
df
```

```python
df.plot(kind='scatter', x = 'reading score', y = 'writing score')
```

```python
df.corr()
```

```python
df[df['writing score'] > 100] # 이상점 확인
```

```python
df.drop(df[df['writing score'] > 100].index, axis = 'index', inplace = True)
print(df[df['writing score'] > 100])
print(df.plot(kind = 'scatter', x = 'reading score', y = 'writing score'))
```

```python
df.corr() # 0.94까지 상승
```

```python
df[(df['writing score'] > 90) & (df['reading score'] < 30)].index
```

```python
df.drop(373, axis = 'index', inplace = True)
print(df.corr())
sns.lmplot(data = df, x = 'reading score', y = 'writing score')
```

```python
df[(df['writing score'] > 90) & (df['reading score'] < 30)]
```

```python
# 실습과제_영화 평점 분석하기 1

df = pd.read_csv('../data/week4/movie_metadata.csv')
df
```

```python
df.plot(kind = 'scatter', x = 'budget', y = 'imdb_score')
```

```python
df['budget'].describe()
```

```python
iqr = df['budget'].quantile(0.75) - df['budget'].quantile(0.25)
iqr
```

```python
condition = df['budget'] > df['budget'].quantile(0.75) + 5*iqr
```

```python
df.drop(df[condition].index, axis = 'index',inplace = True)
df.plot(kind = 'scatter', x = 'budget', y = 'imdb_score')
```

```python
# 실습과제_영화 평점 분석하기 2
df = pd.read_csv('../data/week4/movie_metadata.csv')
budget_head = (df.sort_values(by = 'budget', ascending = False)).head(15)

df.drop(budget_head.index, axis = 'index', inplace = True)
df.plot(kind = 'scatter', x = 'budget', y = 'imdb_score')
```
