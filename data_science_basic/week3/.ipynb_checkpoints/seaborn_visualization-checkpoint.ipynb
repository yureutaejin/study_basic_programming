Seaborn : Statistical Data Visualization (통계를 기반으로한 시각화)  
PDF: Probability Density Function (확률 밀도 함수)  
확률 밀도함수는 데이터셋의 분포를 나타낸다. 특정 구간의 확률은 그래프 아래 그 구간의 면적과 동일하다.  
그래프 아래 모든 면적을 더하면 1이 된다.  
KDE: Kernal Density Estimation

```python
# KDE plot
# !pip install seaborn == 0.9.0

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
body_df = pd.read_csv('../data/week3/body.csv', index_col = 0)

body_df['Height'].value_counts().sort_index() # index 순으로 정렬됨
body_df['Height'].value_counts().sort_index().plot()
body_df
```

```python
sns.kdeplot(body_df['Height']) # 곡선을 부드럽게 만들어줌 (PDF)
sns.kdeplot(body_df['Height'], bw = 1)
```

```python
sns.kdeplot(body_df['Height'], bw = 0.05) # bw로 구간 설정
```

```python
# KDE uses
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns

body_df = pd.read_csv('../data/week3/body.csv', index_col = 0)
body_df.head()
```

```python
body_df.plot(kind = 'hist', y = 'Height', bins = 15)
```

```python
sns.distplot(body_df['Height'], bins = 15) # .distplot => hist 위에 kde(pdf)를 얹음
```

```python
body_df.plot(kind = 'box', y = 'Height')
```

```python
sns.violinplot(y=body_df['Height']) # boxplot을 kde형태로 나타냄 => 가오리 형태라 violin
```

```python
body_df.plot(kind = 'scatter', x = 'Height', y ='Weight')
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight']) # 산점도 그래프를 kde를 통해 등고선 그래프로 나타냄. 3차원. 상관관계와 분포를 동시에 볼 수 있음
```

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```

```python
# 실습과제_교수님의 연봉은?
salary_df = pd.read_csv('../data/week3/salaries.csv')
sns.violinplot(x = salary_df['salary'])
```

```python
# LM plot (Linear Model)
body_df.plot(kind = 'scatter', x = 'Height', y = 'Weight')
```

```python
sns.lmplot(data = body_df, x = 'Height', y = 'Weight') # 회귀선을 그려줌, "data = ~" 꼭 입력해줘야 함
```

```python
# visualization by category 매우 중요!
laptops_df = pd.read_csv('../data/week3/laptops.csv')
laptops_df.head()
laptops_df['os'].unique()
laptops_df
```

```python
sns.catplot(data=laptops_df, x='os', y ='price', kind ='box')
```

```python
sns.catplot(data=laptops_df, x='os', y ='price', kind ='violin')
```

```python
sns.catplot(data=laptops_df, x='os', y ='price', kind ='strip')
```

```python
laptops_df['processor_brand'].unique()
```

```python
sns.catplot(data=laptops_df, x ='os', y='price', kind = 'strip', hue='processor_brand') # hue: hue 분류에 따라 색을 다르게
```

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'swarm', hue = 'processor_brand')
```
