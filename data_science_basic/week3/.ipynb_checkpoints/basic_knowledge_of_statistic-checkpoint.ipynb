```python
# correlation coefficient
%matplotlib inline
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/week3/exam_2.csv')
df.corr()
```

```python
# 상관계수 시각화, 색이 밝을수록 상관계수가 높음
sns.heatmap(df.corr(), annot = True) # annot = True => 숫자도 보여줌

sns.lmplot(data = df, x = 'math score', y = 'reading score')
```
