```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# Normalization by sklearn
# Feature Scaling
nba_player_of_the_week_df = pd.read_csv('../data/NBA_player_of_the_week.csv')
nba_player_of_the_week_df.head()
```

```python
nba_player_of_the_week_df.describe()
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM', 'Weight KG', 'Age']]
height_weight_age_df.head()
```

```python
# Min-Max Scaler
scaler = preprocessing.MinMaxScaler()
normalized_data = scaler.fit_transform(height_weight_age_df)
normalized_df = pd.DataFrame(normalized_data, columns = ['Height', 'Weight', 'Age'])
normalized_df.head()
```

```python
normalized_df.describe()
```

```python
# Standardization

## 소수점 5번째 자리까지만 출력되도록 설정
pd.set_option('display.float_format', lambda x: '%.5f' % x)

scaler = preprocessing.StandardScaler()
standardized_data = scaler.fit_transform(height_weight_age_df)
standardized_df = pd.DataFrame(standardized_data, columns=['Height', 'Weight', 'Age'])
standardized_df.head()
```

```python
standardized_df.describe()
```

```python
# 실습과제_Normalization 직접 해보기
# 필요한 도구 임포트
from sklearn import preprocessing
import pandas as pd

PATIENT_FILE_PATH = '../data/liver_patient_data.csv'
pd.set_option('display.float_format', lambda x: '%.5f' % x)

# 데이터 파일을 pandas dataframe으로 가지고 온다
liver_patients_df = pd.read_csv(PATIENT_FILE_PATH)

# Normalization할 열 이름들
features_to_normalize = ['Total_Bilirubin','Direct_Bilirubin', 'Alkaline_Phosphotase', 'Alamine_Aminotransferase']

# 코드를 쓰세요
scaler = preprocessing.MinMaxScaler()
normalized_data = scaler.fit_transform(liver_patients_df[features_to_normalize])
normalized_df = pd.DataFrame(normalized_data, columns=features_to_normalize)

# 체점용 코드
normalized_df.describe()
```
