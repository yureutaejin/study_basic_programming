```python
import pandas as pd
import numpy as np
```

```python
# From list of lists
two_dimensional_list = [['dongwook', 50 , 86], ['sineui', 89, 31], ['ikjoong', 68, 91], ['yoonsoo', 88, 75]]
df1 = pd.DataFrame(two_dimensional_list, columns = ['name', ' eng_score', 'math_score'], index = ['a', 'b', 'c', 'd'])
print(type(df1))
print(df1.columns)
print(df1.index)
print(df1.dtypes) # column별 자료형 출력
df1
```

```python
# From array of arrays

two_dimensional_array = np.array(two_dimensional_list)
df2 = pd.DataFrame(two_dimensional_array)
df2
```

```python
# From list of series

list_of_series = [
    pd.Series(['dongwook', 50, 86]),
    pd.Series(['sineui', 89, 31]),
    pd.Series(['ikjoong', 68, 91]),
    pd.Series(['yoonsoo', 88, 75])
]

df3 = pd.DataFrame(list_of_series)
df3
```

```python
# From dict of lists, dict of arrays, dict of series

names = ['dongwook', 'sineui', 'ikjoong', 'yoonsoo']
eng_score = [50, 89, 68, 88]
math_score = [86, 31, 91, 75]

dict1 = {
    'name': names,
    'eng_score': eng_score,
    'math_score': math_score
}

dict2 = {
    'name': np.array(names),
    'eng_score': np.array(eng_score),
    'math_score': np.array(math_score)
}

dict3 = {
    'name': pd.Series(names),
    'eng_score': pd.Series(eng_score),
    'math_score': pd.Series(math_score)
}

df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
df3 = pd.DataFrame(dict3)

print(df1)
print(df2)
print(df3)
```

```python
# From list of dicts

my_list = [
    {'name': 'dongwook', 'english_score': 50, 'math_score': 86},
    {'name': 'sineui', 'english_score': 89, 'math_score': 31},
    {'name': 'ikjoong', 'english_score': 68, 'math_score': 91},
    {'name': 'yoonsoo', 'english_score': 88, 'math_score': 75}
]

df = pd.DataFrame(my_list)
df
```

```python
# 실습과제_스타들의 생일은 언제?
name = ["Taylor Swift", 'Aaron Sorkin', 'Harry Potter', 'Ji-Sung Park']
birthday = ['December 13, 1989', "June 9, 1961", "July 31, 1980", "February 25, 1981"]
occupation = ['Singer-songwriter', 'Screenwriter', 'Wizard', 'Footballer']

dict_star = {
    'name': name,
    'birthday': birthday,
    'occupation': occupation
}

df = pd.DataFrame(dict_star)
df
```

# data types of pandas  
``df.dtypes``로 확인  
한 colunm 내에서는 모든 값이 동일한 데이터 타입  

**int64** = 정수  
**float64** = 실수  
**object** = 텍스트  
**bool** = boolean  
**datetime64** = 날짜와 시간  
**category** = 카테고리
