```python
# 실습과제_TV 시청률 크롤링 1
## 주소: https://workey.codeit.kr/ratings/index

import requests

rating_page = requests.get('https://workey.codeit.kr/ratings/index') # requests.get("주소")
print(rating_page.text) # print를 사용해야 텍스트 문법이 적용됨
```

```python
# 실습과제_TV 시청률 크롤링 2
import requests

rating_pages = []

year = 2010
month = 1
week_index = 0

for y in range(year, 2013, 1):
    month = 1
    for m in range(month, 13, 1):
        week_index = 0
        for w in range(week_index, 5, 1):
            page = requests.get('https://workey.codeit.kr/ratings/index?year={}&month={}&weekIndex={}'.format(year, month, week_index))
            rating_pages.append(page.text)
        
print(len(rating_pages))
print(rating_pages[0])
```
