```python
# 실습과제_타이타닉 EDA
df = pd.read_csv('../data/week3/titanic.csv')
df
```

```python
## 타이타닉의 승객은 30대와 40대가 가장 많다. => x
print(df['Age'].plot(kind = 'hist', bins = 10))

print(sns.kdeplot(df['Age']))
```

```python
## 가장 높은 요금을 낸 사람은 30대이다. => o
print(df.plot(x = 'Age', y = 'Fare', kind = 'scatter'))
# or
df['Fare'].sort_values(ascending = False)
df.loc[679,]
```

```python
## 생존자가 사망자보다 많다 => x
df['Survived'].value_counts()
```

```python
## 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다. => o
df['Pclass'].value_counts()
```

```python
## 가장 생존율이 높은 객실 등급은 1등실이다. => o
np.sum(df.isna())
df.plot(kind = 'scatter', x = 'Pclass', y = 'Survived')
sns.kdeplot(kind = 'scatter', x = df['Pclass'], y = df['Survived'])

# or
print(df.groupby('Pclass').mean())
```

```python
## 나이가 어릴수록 생존율이 높다. => x
sns.catplot(data = df, kind = 'strip', x = 'Survived', y = 'Age')
print(sns.catplot(data= df, kind = 'violin', x = 'Survived', y ='Age'))

# or
df.corr()[['Survived', 'Age']]
```

```python
# or
df.groupby('Survived').plot(kind='hist', y='Age')
```

```python
# or
print(sns.kdeplot(x = df['Survived'], y = df['Age'].fillna(-99)))
```

```python
## 나이보다 성별이 생존율에 더 많은 영향을 미친다. => o
sns.catplot(data = df, kind = 'strip', x = 'Survived', y = 'Age', hue = 'Sex')
```
