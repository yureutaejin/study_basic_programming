```python
# 실습과제_직업 탐구하기1
df = pd.read_csv('../data/week3/occupations.csv')
group_occupation = df.groupby('occupation')
pd.DataFrame((group_occupation['age'].mean()).sort_values(ascending = True))
```

```python
# 실습과제_직업 탐구하기 2
df = pd.read_csv('../data/week3/occupations.csv')
group_occupation_female = df[df['gender'] == 'F'].groupby('occupation')
group_occupation = df.groupby('occupation')
result = group_occupation_female.count() / group_occupation.count()
result = (result.sort_values(by = 'gender', ascending = False))['gender']
result.loc['doctor'] = 0
result
```

```python
# merge data
price_df = pd.read_csv('../data/week3/vegetable_price.csv')
quantity_df = pd.read_csv('../data/week3/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on= 'Product') # Inner Join
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how ='left') # Left Outer Join
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how = 'right') #Right Outer Join
```

```python
pd.merge(price_df, quantity_df, on = 'Product', how = 'outer') #Full Outer Join
```
