```python
# 실습과제_실리콘 밸리에는 누가 일할까?2
%matplotlib inline
import pandas as pd

df = pd.read_csv('../data/week3/silicon_valley_details.csv')

adobe = (df['company'] == 'Adobe') & (df['race'] == 'Overall_totals') & (df['count'] != 0)
except_total = (df['job_category'] != 'Totals') & (df['job_category'] != 'Previous_totals')
adobe_job = df[adobe & except_total]
(adobe_job.set_index('job_category')).plot(kind = 'pie', y ='count')
```

```python
# 실습과제_국가 지표 분석하기

df = pd.read_csv('../data/week3/world_indexes.csv')

# 질문 1 가장 연관성 깊은 지표 찾기
indicators = ['Life expectancy at birth- years', 'Internet users percentage of population 2014', 'Forest area percentage of total land area 2012', 'Carbon dioxide emissionsAverage annual growth']


for i in range(len(indicators)):
    j = i + 1
    while j != len(indicators):
        print(df.plot(kind = 'scatter', x = indicators[i] , y = indicators[j]))
        print(df[[indicators[i], indicators[j]]].corr())
        print()
        j += 1
```
