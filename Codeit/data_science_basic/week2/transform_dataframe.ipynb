```python
iphone_df = pd.read_csv('../data/week2/iphone.csv', index_col = 0)
# append or remove value at DataFrame

# append
iphone_df.loc['iPhone XR'] = ['2018-10-26', 6.1, '3GB', 'iOS 12.0.1', 'Yes']
iphone_df['제조사'] = 'Apple'
iphone_df

# index, column remove
iphone_df.drop('iPhone XR', axis = 'index', inplace = False) # index에서 iPhone XR을 지운 df를 만들어라(False -> 새 df를 만듦)
iphone_df.drop('제조사', axis = 'columns', inplace = True) # df에서 column에서 제조사를 지워라
iphone_df.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis = 'index', inplace = False)
```

```python
# 실습과제_잘못된 DataFrame 고치기
df = pd.read_csv('../data/week2/body_imperial1.csv', index_col = 0)
df.loc[1, 'Weight (Pound)'] = 200 # df.iloc[1, 1] = 200
df.drop(21, axis = 'index', inplace = True)
df.loc[20] = [70, 200]
df
```

```python
# 실습과제_데이터 변형하기
df['비만도'] = '정상'
df.loc[:10, 'Gender'] = 'Male'
df.loc[11:, 'Gender'] = 'Female' # 정수 기반(iloc) 슬라이싱: a:b -> a ~ b-1, label 기반(loc): a:b -> a~b
df
```

```python
# set up index and columns
liverpool_df = pd.read_csv('../data/week2/liverpool.csv', index_col = 0)
liverpool_df.rename(columns={'position' : 'Position'}, inplace = True) # rename 메소드 -> .rename(index or columns = {a:b,,,....}, inplace = True or False)
liverpool_df.rename(columns={'position' : 'Position', 'born': 'Born', 'number': 'Number', 'nationality':'Nationality'}, inplace = True)


liverpool_df.index.name = 'Player Name' # index 항목 전체 분류 naming

# index 바꾸기
liverpool_df['Player Name'] = liverpool_df.index # 먼저 columns로 이동시키지 않으면 데이터가 날라감
liverpool_df.set_index('Number', inplace = True)
liverpool_df
```

```python
# 실습과제_서류 전형 합격 여부
df = pd.read_csv('../data/week2/toeic.csv')
pass_check = ((df['LC'] + df['RC']) >= 600) &((df['LC'] >= 250) & (df['RC'] >= 250))
df['합격 여부'] = pass_check
df
```
